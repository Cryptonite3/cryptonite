```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from itables import init_notebook_mode
init_notebook_mode(all_interactive=True)
```

```python
df = pd.read_csv('../Data/clean_df.csv', index_col = [0, 1])
```

# Correlations

Firstly, we can try to look as the correlation matrix as a heatmap which shows us which of our attributes have correlations between each other and the magnitude of the same.

```python
plt.figure(figsize = (10, 10))
ax = sns.heatmap(data = df.corr(), cmap = 'Blues')
ax.set(xlabel = 'Attribute', ylabel = 'Attribute')
ax.set_title('Correlation Heatmap')
plt.show()
```

From the above heatmap, we can see that 24h and 7d turn out to have the highest or actually any correlation(about 0.65) for that matter.  
We see a bit of color towards the bottom right of the plot implying that the traded amounts between time frames have a bit of correlation between each other as practically they are an aggregate of each other.

# Box Plot for Price

Now we can try to have a look at the price variations of all the cryptocurrencies we have.

```python
plt.figure(figsize = (10, 4))
ax = sns.boxplot(x = df['price'])
ax.set_title('Box Plot for Price')
ax.set(xlabel = 'Price', ylabel = 'CryptoCurrency')
plt.show()
```

We can clearly notice class imbalance in this boxplot implying that not all cryptocurrencies are successful and only a few manage to climb up on the price ladder.  
Only 3 of them are high enough that they are noticed differently or superior to the others.

# Histogram for 24h

A histogram would show us densities within the 24h data of our dataset.

```python
ax = sns.histplot(x = df['24h'])
ax.set_title('Histogram for past 24h stock prices')
ax.set(xlabel = '24h')
plt.show()
```

The above plot shows us where the 24h data lies for most cryptocurrencies.
